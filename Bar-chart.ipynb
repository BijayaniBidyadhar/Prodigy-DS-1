```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
from matplotlib import style
```

```python
data = pd.read_csv("Population-of-India.csv")
```

```python
data.head()
```

```python
data = data[data['State/UT'] != 'Total (India)']
```

```python
plt.figure(figsize=(15, 8))
plt.bar(data['State/UT'], data['Population[50]'], color='#151E3D')
plt.xlabel('State/UT', fontsize=20)
plt.ylabel('Population', fontsize=20)
plt.title('State-wise Population in India', fontsize=20)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create bar chart for male population
plt.figure(figsize=(20, 15))
bar_width = 0.4
bar_positions = range(len(data))

m_bar = (data['Male'] / data['Population[50]']) * 100
m_bar=plt.bar(data['State/UT'], data['Male'], color='#1F456E', label='Male Population')
plt.xlabel('State/UT', fontsize=20)
plt.ylabel('Population', fontsize=20)
plt.title('State-wise Male Population in India', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(data['State/UT'], rotation=45, ha='right', fontsize=15)
# plt.text(m_bar.get_x() + boys_bar.get_width() / 2, boys_percentage_value + 1, f'{boys_percentage_value:.1f}%',
#              ha='center', va='bottom', color='blue', fontsize=9)
plt.legend()
plt.tight_layout()

# Show the plot for male population
plt.show()
```

```python
# Create bar chart for male population
plt.figure(figsize=(20, 15))
bar_width = 0.4
bar_positions = range(len(data))

f_bar = (data['Female'] / data['Population[50]']) * 100
f_bar=plt.bar(data['State/UT'], data['Female'], color='#48AAAD', label='Female Population')
plt.xlabel('State/UT', fontsize=20)
plt.ylabel('Population', fontsize=20)
plt.title('State-wise Female Population in India', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(data['State/UT'], rotation=45, ha='right', fontsize=15)
# plt.text(m_bar.get_x() + boys_bar.get_width() / 2, boys_percentage_value + 1, f'{boys_percentage_value:.1f}%',
#              ha='center', va='bottom', color='blue', fontsize=9)
plt.legend()
plt.tight_layout()

# Show the plot for male population
plt.show()
```

```python
# Calculate the total population for each state
data['Total Population'] = data['Male'] + data['Female']
plt.figure(figsize=(15, 9))
bar_width = 0.4
bar_positions = range(len(data))

m_percentage = (data['Male'] / data['Total Population']) * 100
f_percentage = (data['Female'] / data['Total Population']) * 100
m_bars = plt.bar(bar_positions, m_percentage, width=bar_width, label='Male', color='#1520A6')
f_bars = plt.bar([pos + bar_width for pos in bar_positions], f_percentage, width=bar_width, label='Female', color='#52B2BF')

# Add percentage values above each bar
for m_bar, f_bar in zip(m_bars, f_bars):
    m_percentage_value = m_bar.get_height()
    f_percentage_value = f_bar.get_height()
       
# Add data labels and ticks
plt.xlabel('State/UT', fontsize=15)
plt.ylabel('Population Percentage', fontsize=15)
plt.title(' Comparition Between Male & Female by State/UT', fontsize=14)
plt.xticks([pos + bar_width / 2 for pos in bar_positions], data['State/UT'], rotation=45, ha='right', fontsize=10)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Calculate the absolute difference between male and female populations
data['Male-Female Difference'] = abs(data['Male'] - data['Female'])

# Sort the DataFrame based on the difference column in descending order
sorted_data = data.sort_values(by='Male-Female Difference', ascending=False).head(15)

# Create a bar chart for the top 10 states with the highest difference
plt.figure(figsize=(15, 8))
bars = plt.barh(sorted_data['State/UT'], sorted_data['Male-Female Difference'], color='#69306d')
for bar in bars:
    yval = bar.get_y() + bar.get_height() / 2
    xval = bar.get_width()
    
plt.xlabel('Male-Female Difference in Population', fontsize=20)
plt.ylabel('State/UT', fontsize=20)
plt.title('Top 10 States/UTs with Highest Difference in Male and Female Population', fontsize=20)
plt.show()
```
